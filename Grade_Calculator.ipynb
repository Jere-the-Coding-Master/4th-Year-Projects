```python
student_name = input("Enter your Name: ") # student name
course_name = input("Enter your Course Name: ") # course name

prelim = float(input("Enter your Prelim Grade: ")) # prelim
midterm = float(input("Enter your Midterm Grade: ")) # midterm
finals = float(input("Enter your Finals Grade: ")) # final

sem_grade = (prelim * 0.3) + (midterm * 0.3) + (finals * 0.4) # sem grade

print("\nStudent Name: {}\
\nCourse Name: {}\
\n\n\tPrelim Grade: \t{}\
\n\tMidterm Grade: \t{}\
\n\tFinals Grade: \t{}\
\n\n\tSEMESTER GRADE: {:.2f}".format(student_name, course_name, prelim, midterm, finals, sem_grade))


# emotion
def emoticon(sem_grade):
  happy, lol, sad = "\U0001F600", "\U0001F606", "\U0001F62D"
  if sem_grade > 70.00:
      return happy
  elif sem_grade == 70.00:
      return lol
  else:
      return sad

print("\nMy emotion for you is ",emoticon(sem_grade))


#equivalent grade
def equal_grade(sem_grade):
  if sem_grade >= 97.00 and sem_grade <= 100:
    gpe = "1.00"
    le = "A"
    d = "Excellent"
    return gpe, le, d

  elif sem_grade >= 93.00 and sem_grade <= 96.99:
    gpe = "1.25"
    le = "A-"
    d = "Superior"
    return gpe, le, d

  elif sem_grade >= 89.00 and sem_grade <= 92.99:
    gpe = "1.50"
    le = "B+"
    d = "Superior"
    return gpe, le, d

  elif sem_grade >= 85.00 and sem_grade <= 88.99:
    gpe = "1.75"
    le = "B"
    d = "Average"
    return gpe, le, d

  elif sem_grade >= 82.00 and sem_grade <= 84.99:
    gpe = "2.00"
    le = "B-"
    d = "Average"
    return gpe, le, d

  elif sem_grade >= 79.00 and sem_grade <= 81.99:
    gpe = "2.25"
    le = "C+"
    d = "Average"
    return gpe, le, d

  elif sem_grade >= 76.00 and sem_grade <= 78.99:
    gpe = "2.50"
    le = "C"
    d = "Passed"
    return gpe, le, d

  elif sem_grade >= 73.00 and sem_grade <= 75.99:
    gpe = "2.75"
    le = "C-"
    d = "Passed"
    return gpe, le, d

  elif sem_grade >= 70.00 and sem_grade <= 72.997:
    gpe = "3.00"
    le = "D"
    d = "Passed"
    return gpe, le, d

  else:
    gpe = "5.00"
    le = "F"
    d = "Failed"
    return gpe, le, d


while True:
  show_grade = input("\nDo you want to see your equivalent grade?(Y/N)")

  if show_grade == "Y":
      what_grade = input("Which of the following do you want to see?\
      \n\t 1. Grade Point Equivalent\
      \n\t 2. Letter Equivalent\
      \n\t 3. Description\n\t ")

      gpe, le, d = equal_grade(sem_grade)

      if what_grade == "1":
          print("\n\tYour Grade Point Equivalent is ", gpe, "\n")
          break
      elif what_grade == "2":
          print("\n\tYour Letter Equivalent is ", le, "\n")
          break
      elif what_grade == "3":
          print("\n\tYour Description is ", d, "\n")
          break
      else:
          print("\nWrong Input")

  elif show_grade == "N":
    print("Thank you so much.")
    break

  else:
    print("Wrong input.\n")


```
